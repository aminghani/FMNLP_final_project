```python
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_name = "Salesforce/blip-image-captioning-base"
processor = BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name)
model.to(device)
```

```python
from PIL import Image
import os
from tqdm import tqdm
from nltk.translate.bleu_score import sentence_bleu
```

```python
funny_bleu = 0
Romantic_bleu = 0

for i in tqdm(range(len(flicker_test))):
    image_name = flicker_test.iloc[i, 0]
    image_path = os.path.join('/content/content/FMNLP_dataset/flickerstyle/images', image_name)
    image = Image.open(image_path)
    text_funny = 'Funny description: '
    text_romantic = 'Romantic description: '
    text_funny += flicker_test.iloc[i, 4]
    text_romantic += flicker_test.iloc[i, 4]
    input_funny = processor(text=text_funny, images=image, return_tensors="pt", truncation=True)
    input_romantic = processor(text=text_romantic, images=image, return_tensors="pt", truncation=True)
    input_funny = {key: val.to(device) for key, val in input_funny.items()}
    input_romantic = {key: val.to(device) for key, val in input_romantic.items()}
    outputs_funny = model.generate(**input_funny)
    outputs_romantic = model.generate(**input_romantic)

    decoded_output_funny = processor.decode(outputs_funny[0], skip_special_tokens=True)
    decoded_output_romantic = processor.decode(outputs_romantic[0], skip_special_tokens=True)

    funny_ref = flicker_test.iloc[i, 1]
    romantic_ref = flicker_test.iloc[i, 2]

    funny_bleu += sentence_bleu([funny_ref.split()], decoded_output_funny.split())
    Romantic_bleu += sentence_bleu([romantic_ref.split()], decoded_output_romantic.split())

```

```python
print(f'bleu on funny description: {funny_bleu / len(flicker_test)}, bleu on romantic: {Romantic_bleu / len(flicker_test)}')
```

```python
from rouge import Rouge
rouge = Rouge()
```

```python
rouge_1_score_fu = 0
rouge_2_score_fu = 0
rouge_l_score_fu = 0

rouge_1_score_ro = 0
rouge_2_score_ro = 0
rouge_l_score_ro = 0


for i in tqdm(range(len(flicker_test))):
    image_name = flicker_test.iloc[i, 0]
    image_path = os.path.join('/content/content/FMNLP_dataset/flickerstyle/images', image_name)
    image = Image.open(image_path)
    text_funny = 'Funny description: '
    text_romantic = 'Romantic description: '
    text_funny += flicker_test.iloc[i, 4]
    text_romantic += flicker_test.iloc[i, 4]
    input_funny = processor(text=text_funny, images=image, return_tensors="pt", truncation=True)
    input_romantic = processor(text=text_romantic, images=image, return_tensors="pt", truncation=True)
    input_funny = {key: val.to(device) for key, val in input_funny.items()}
    input_romantic = {key: val.to(device) for key, val in input_romantic.items()}
    outputs_funny = model.generate(**input_funny)
    outputs_romantic = model.generate(**input_romantic)

    decoded_output_funny = processor.decode(outputs_funny[0], skip_special_tokens=True)
    decoded_output_romantic = processor.decode(outputs_romantic[0], skip_special_tokens=True)

    funny_ref = flicker_test.iloc[i, 1]
    romantic_ref = flicker_test.iloc[i, 2]

    scores_funny = rouge.get_scores(decoded_output_funny, funny_ref)
    scores_romantic = rouge.get_scores(decoded_output_romantic, romantic_ref)

    rouge_1_score_fu += scores_funny[0]['rouge-1']['f']
    rouge_2_score_fu += scores_funny[0]['rouge-2']['f']
    rouge_l_score_fu += scores_funny[0]['rouge-l']['f']

    rouge_1_score_ro += scores_romantic[0]['rouge-1']['f']
    rouge_2_score_ro += scores_romantic[0]['rouge-2']['f']
    rouge_l_score_ro += scores_romantic[0]['rouge-l']['f']
```

```python
print(f'funny: r1: {rouge_1_score_fu / len(flicker_test)}, r2: {rouge_2_score_fu / len(flicker_test)}, rl: {rouge_l_score_fu / len(flicker_test)}')
print(f'romantic: r1: {rouge_1_score_ro / len(flicker_test)}, r2: {rouge_2_score_ro / len(flicker_test)}, rl: {rouge_l_score_ro / len(flicker_test)}')
```

```python
import nltk
from nltk.translate import meteor_score
nltk.download('wordnet')
```

```python
def calculate_meteor_score(reference, hypothesis):
    reference_tokens = reference.split()
    hypothesis_tokens = hypothesis.split()

    score = meteor_score.meteor_score([reference_tokens], hypothesis_tokens)

    return score
```

```python
meteor_fu = 0

meteor_ro = 0

for i in tqdm(range(len(flicker_test))):
    image_name = flicker_test.iloc[i, 0]
    image_path = os.path.join('/content/content/FMNLP_dataset/flickerstyle/images', image_name)
    image = Image.open(image_path)
    text_funny = 'Funny description: '
    text_romantic = 'Romantic description: '
    text_funny += flicker_test.iloc[i, 4]
    text_romantic += flicker_test.iloc[i, 4]
    input_funny = processor(text=text_funny, images=image, return_tensors="pt", truncation=True)
    input_romantic = processor(text=text_romantic, images=image, return_tensors="pt", truncation=True)
    input_funny = {key: val.to(device) for key, val in input_funny.items()}
    input_romantic = {key: val.to(device) for key, val in input_romantic.items()}
    outputs_funny = model.generate(**input_funny)
    outputs_romantic = model.generate(**input_romantic)

    decoded_output_funny = processor.decode(outputs_funny[0], skip_special_tokens=True)
    decoded_output_romantic = processor.decode(outputs_romantic[0], skip_special_tokens=True)

    funny_ref = flicker_test.iloc[i, 1]
    romantic_ref = flicker_test.iloc[i, 2]

    meteor_fu += calculate_meteor_score(funny_ref, decoded_output_funny)

    meteor_ro += calculate_meteor_score(romantic_ref, decoded_output_romantic)
```

```python
print(f'funny meteor: {meteor_fu / len(flicker_test)}, romantic meteor: {meteor_ro / len(flicker_test)}')
```

```python
bleu = 0
model.to(device)
mask = ~test_senti_cap['filename'].duplicated(keep='first')
test_senti_cap_ = test_senti_cap[mask]

for i in tqdm(range(len(test_senti_cap_))):
    image_name = test_senti_cap_.iloc[i, 3]
    image_path = os.path.join('/content/content/FMNLP_dataset/senticap_images', image_name)
    try:
        image = Image.open(image_path)
    except:
        continue
    senti = int(test_senti_cap_.iloc[i, 7])
    text = ''
    if senti == 1:
        text += 'positive caption: '
    else:
        text += 'negative caption: '
    input = processor(text=text, images=image, return_tensors="pt", truncation=True)
    input = {key: val.to(device) for key, val in input.items()}
    outputs = model.generate(**input)

    decoded_output = processor.decode(outputs[0], skip_special_tokens=True)

    ref = test_senti_cap_.iloc[i, 8]

    bleu += sentence_bleu([ref.split()], decoded_output.split())

```

```python
print(bleu / len(test_senti_cap_))
```

```python
rouge_1_score = 0
rouge_2_score = 0
rouge_l_score = 0

for i in tqdm(range(len(test_senti_cap_))):
    image_name = test_senti_cap_.iloc[i, 3]
    image_path = os.path.join('/content/content/FMNLP_dataset/senticap_images', image_name)
    try:
        image = Image.open(image_path)
    except:
        continue
    senti = int(test_senti_cap_.iloc[i, 7])
    text = ''
    if senti == 1:
        text += 'positive caption: '
    else:
        text += 'negative caption: '
    input = processor(text=text, images=image, return_tensors="pt", truncation=True)
    input = {key: val.to(device) for key, val in input.items()}
    outputs = model.generate(**input)

    decoded_output = processor.decode(outputs[0], skip_special_tokens=True)

    ref = test_senti_cap_.iloc[i, 8]

    scores = rouge.get_scores(decoded_output, ref)

    rouge_1_score += scores[0]['rouge-1']['f']
    rouge_2_score += scores[0]['rouge-2']['f']
    rouge_l_score += scores[0]['rouge-l']['f']
```
